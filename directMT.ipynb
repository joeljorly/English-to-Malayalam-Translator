```
#!/usr/bin/env python
# coding: utf-8

# In[1]:


import pandas as pd
import string


# In[2]:


st=pd.read_csv('mdata.csv',delimiter='\t')


# In[3]:


englist=st['english_word']
mallist=st['malayalam_definition']


# In[4]:


class translation:
    def __init__(self,input):
        self.input=input
        
        self.translate()
    
    def remove_stop(self,word_list):
        stop_words=['Is','To','Are','-','.',':',';','/']
        new_list=[]
        for word in word_list:
            if word not in stop_words:
                new_list.append(word)
        return new_list
    def translate(self):
        str=string.capwords(self.input)
        str=str.split(" ")
        str=self.remove_stop(str)
        for word in str:
            c=0
            for k in englist:
                c=c+1
                if word==k:
                    print(mallist[c-1],end=' ')
                    break


# In[5]:


input='he is a good boy'
translation(input)


# In[ ]:




```
